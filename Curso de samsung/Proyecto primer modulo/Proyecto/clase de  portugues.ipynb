```python
#valor maximo de ausencias

ab=nb2['absences'].max()
print(ab)
```

```python
#porcentaje de asistencia.

asistir=[]
maximo= nb2['absences'].max()
for y in nb2['absences']:
    asistir.append(y)
print(asistir)
print(maximo)


asistencias=[]
for a in range (0,len(nb2.index)):
    inasistencias=(-((asistir[a]/maximo)*100)+100)
    asistencias.append(inasistencias)
print(asistencias)
print(len(asistencias))
```

```python
#Porcentaje de asistencias.

c=1
for r in asistencias:
    print(f'Estudinte {c}... su porcentage de asistencia fue del {(round(r,2))}%')
    c+=1
```

```python
#Notas Finales 
Primera_Nota=[]
Segunda_Nota=[]
Tecera_Nota=[]

for i in nb2['G1']:
    Primera_Nota.append(i)
    for u in nb2['G2']:
        Segunda_Nota.append(u)
        for l in nb2['G3']:
                Tecera_Nota.append(l)
```

```python
NT=[]
for o in range (0,len(nb2.index)):
    notafinal=(Primera_Nota[o]+Segunda_Nota[o]+Tecera_Nota[o])/3
    NT.append(notafinal)
print(NT)
```

```python
#Aprovados y desaprobados.

ex=[]
app=[]
for L in range(0,len(NT)):
    A=asistencias[L]
    N=NT[L]
    if(A < 80):
        app.append(0)
        ex.append(0)
    elif (A > 80 and N < 10):
        app.append(0)
        ex.append(0)
    elif(A >80 and N >= 10 and N <= 15):
        app.append(1)
        ex.append(1)
    elif(A > 80 and N > 15):
        app.append(1)
        ex.append(0)
print(ex)
print(app)
```

```python
# crear las dcolumnas de approved y extra.

nb2.insert(16,'approved',app,allow_duplicates=False)
nb2.insert(17,'extra',ex,allow_duplicates=False)
```
